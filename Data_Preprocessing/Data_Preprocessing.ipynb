```python
#Position / PreferredFoot / AttackingWorkRate / DefensiveWorkRate

from sklearn.preprocessing import LabelEncoder

class_le = LabelEncoder()
y = class_le.fit_transform(df['Position'].values)
x = class_le.transform(df1['Position'].values)
df['Position']=y
df1['Position']=x

class_le1 = LabelEncoder()
y = class_le1.fit_transform(df['PreferredFoot'].values)
x = class_le1.transform(df1['PreferredFoot'].values)
df['PreferredFoot']=y
df1['PreferredFoot']=x

class_le2 = LabelEncoder()
y = class_le2.fit_transform(df['AttackingWorkRate'].values)
x = class_le2.transform(df1['AttackingWorkRate'].values)
df['AttackingWorkRate']=y
df1['AttackingWorkRate']=x

class_le3 = LabelEncoder()
y = class_le3.fit_transform(df['DefensiveWorkRate'].values)
x = class_le3.transform(df1['DefensiveWorkRate'].values)
df['DefensiveWorkRate']=y
df1['DefensiveWorkRate']=x

df.head()
```

## Identification of Correlations between Independent Variables -> Threshold : 0.7

```python
feature_sel = []
for i in df.columns:
    for a in df.columns :
        if abs(df[i].corr(df[a]))>0.7:
            if i != a:
                print(f"{i} & {a} : {df[i].corr(df[a]):.2f}")
                feature_sel.append([i, a])
```

```python
feature_sel
```

## Removed Some Features beyond the threshold

```python
remove_list = []

for i in range(len(feature_sel)):
    if 'Crossing' in feature_sel[i][0]:
        df.drop(feature_sel[i][1], axis=1, inplace=True)
        df1.drop(feature_sel[i][1], axis=1, inplace=True)
        remove_list.append(feature_sel[i])
        
for i in range(len(remove_list)):
    feature_sel.remove(remove_list[i])
    
print(df.columns)
print()
print(feature_sel)
```
